```python
import os

import numpy as np
import torch

from torchvision import transforms

from PIL import Image

from importlib import reload

import matplotlib.pyplot as plt
```

```python
def show_reconstruction(path, idx):
    original = Image.open(path + f'/ori_{idx}.jpg')
    recon = Image.open(path + f'/rec_{idx}.jpg')
    
    fig, axes = plt.subplots(1, 2, figsize=(8, 4))
    
    axes[0].imshow(original)
    axes[1].imshow(recon)
    
    axes[0].axis('off')
    axes[1].axis('off')
    
    axes[0].set_title("original")
    axes[1].set_title("reconstruction")
    plt.show()
    
    
folder = "./ATSPrivacy-Framework/benchmark/images/data_cifar100_arch_ResNet20-4_epoch_200_optim_inversed_mode_normal_auglist__rlabel_False"

for i in range(1, 50):
    show_reconstruction(folder, i)
```

```python
def show_multiple_reconstructions(result_dictionary, idx):    
    imgsize = 4
    fig, axes = plt.subplots(1, len(result_dictionary), figsize=(len(result_dictionary)*imgsize, imgsize))
    
    for i, (title, prefix) in enumerate(result_dictionary.items()):
        if title == "original":
            img = Image.open(prefix + f'_{idx}.jpg')
        else:
            img = Image.open(prefix + f'_{idx}.jpg')
            
        axes[i].imshow(img)
        axes[i].axis('off')
        axes[i].set_title(title)
        
    axes[0].set_ylabel(f"image {idx}")
        
    plt.show()
        
# {title : prefix}
result_dictionary = {
    "original" : r"./ATSPrivacy-Framework/benchmark/images/data_cifar100_arch_ResNet20-4_epoch_200_optim_inversed_mode_normal_auglist__rlabel_False/ori",
    "3-1-7 default defense" : r"./ATSPrivacy-Framework/benchmark/images/data_cifar100_arch_ResNet20-4_epoch_200_optim_inversed_mode_aug_auglist_3-1-7_rlabel_False/ori",
    "3-1-7 default attack" : r"./ATSPrivacy-Framework/benchmark/images/data_cifar100_arch_ResNet20-4_epoch_200_optim_inversed_mode_aug_auglist_3-1-7_rlabel_False/rec",
    "3-1-7 defense shifted" : r".\ATSRefactored\results\images\data_cifar100_arch_ResNet20-4_epoch_200_optim_inversed_mode_aug_auglist_3-1-7_rlabel_False_reaugment_translate_clipped1/ori_isolated",
    "3-1-7 intelligent attack shifted" : r".\ATSRefactored\results\images\data_cifar100_arch_ResNet20-4_epoch_200_optim_inversed_mode_aug_auglist_3-1-7_rlabel_False_reaugment_translate_clipped1/rec_isolated",
    "3-1-7 intelligent attack" : r".\ATSRefactored\results\images\data_cifar100_arch_ResNet20-4_epoch_200_optim_inversed_mode_aug_auglist_3-1-7_rlabel_False_reaugment_translate_clipped1/rec_untransformed",
}

for i in range(1, 11):
    show_multiple_reconstructions(result_dictionary, i)
```
